```python
#Getting percentage returns of buyhold strategy
df['prices_BuyHold']=(df['prices'].pct_change(1)+1).cumprod()

def Strat(maBig,maSmall,stdv,cond,npos): #stdv = standard deviation of the last __ days
    
    df['maBig']=df['prices'].rolling(maBig).mean()
    df['maSmall']=df['prices'].rolling(maSmall).mean()


    #Getting percentage returns of moving averages

    #coping with no MA data for the first maBig days
    position = []
    for i in range (maBig):
        position.append(1)
    
    #moving average strategy
    for i in range (timeRange-maBig):
        if (df.iloc[i+maBig,2]) < (df.iloc[i+maBig,3]):    #if big ma is smaller than small ma, we buy and v.v.
            position.append(1)
        else:
            position.append(0)
    
    #Standard Deviations
    placeholder = []
    recentStandardDeviations = []
    c = 0
    for i in range(stdv):
        recentStandardDeviations.append(0)
    for i in range(timeRange - stdv):
        for j in range(stdv):
            placeholder.append(df.iloc[j+c,0])
        recentStandardDeviation = statistics.stdev(placeholder)
        recentStandardDeviations.append(recentStandardDeviation)
        c += 1
        placeholder.clear()
        
    df['recentStDv'] = recentStandardDeviations   
    
    for i in range(timeRange):
        if recentStandardDeviations[i] >=cond:
            position[i]=npos
    
    df['position'] = position
        
    #ensure delay as we cannot buy after close
    df['position']=df['position'].shift()

    #calculate the daily percentage returns of strategy
    df['strategyPct']=df['prices'].pct_change(1)*df['position']

    #calculate cummulative returns using strategy
    df['strategy']=(df['strategyPct']+1).cumprod()

#print(df)
```
